```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
```

```python
#The amount of contamination of the data set, i.e. the proportion of outliers in the data set. Used when fitting to define the threshold on the scores of the samples.
from sklearn.ensemble import IsolationForest
clf = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)
clf.fit(X)
```

```python
y_pred = clf.predict(X)
```

```python
#sequential color maps ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='plasma')
plt.title("Isolation Forest Outlier Detection on DIGITS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = clf.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='magma')
plt.title("Local Outlier Factor Outlier Detection on DIGITS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
# Load the DIGITS dataset
digits = load_digits()
X = digits.data
y = digits.target
```

```python
# Fit Isolation Forest model
clf_iso = IsolationForest(contamination=0.1, random_state=42)
y_pred_iso = clf_iso.fit_predict(X)
```

```python
# Fit Local Outlier Factor model
clf_lof = LocalOutlierFactor(contamination=0.1)
y_pred_lof = clf_lof.fit_predict(X)
```

```python
# Plot Isolation Forest outliers
plt.scatter(X[:, 0], X[:, 1], c=np.where(y_pred_iso == -1, 'orange', 'black'), label='Isolation Forest')
plt.title("Outlier Detection using Isolation Forest on DIGITS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
# Plot Local Outlier Factor outliers
plt.scatter(X[:, 0], X[:, 1], c=np.where(y_pred_lof == -1, 'orange', 'black'), label='Local Outlier Factor')
plt.title("Outlier Detection using Local Outlier Factor on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```
